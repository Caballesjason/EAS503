```python
# library import
from utility_functions import *
import plotly.graph_objects as go
import pandas as pd
import ipywidgets as widgets
from plotly.offline import iplot
import plotly
```

```python
# database connection
database = 'netflix.db'
conn = create_connection(database)
```

```python
"""
--- Importing the Data ---
# We want to have two datasets
"""

sql = """
SELECT
	 TITLE,
	 FILM_TYPE, 
	 RATING, 
	 DESCRIPTION, 
 Films_And_Countries.COUNTRIES,
 Films_And_Genres.GENRE
 FROM Films
 INNER JOIN Films_And_Genres ON Films.FILM_ID = Films_And_Genres.FILM_ID
 INNER JOIN Films_And_Countries ON Films.FILM_ID = Films_And_Countries.FILM_ID
 WHERE Films_And_Genres.GENRE = 'TV Dramas'
"""

df = pd.read_sql(sql, conn)
```

```python
# "" Indicates there were was no month avaiable in original data
# Ask if countries should be accounted for each genre they are labelled as

fig = go.Figure(data=[go.Table(
    header=dict(values=list(df.columns),
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[df.TITLE, df.FILM_TYPE, df.RATING, df.DESCRIPTION, df.COUNTRIES, df.GENRE],
               fill_color='lavender',
               align='left'))
])

fig.update_layout(
    title={
        'text': "Films Categorized as TV Dramas",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

fig.show()
```
